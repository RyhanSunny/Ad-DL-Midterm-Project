# Model selection

## Implementation of various TensorFlow models on fashion minst dataset without using any optimizers.

```python
from fashionmnist_model import FMM
```

```python
# Load the data
X_train, y_train, X_test, y_test = FMM.load_data()
```

## Using the first version

```python
model = FMM.create_model_v1()
history = FMM.compile_and_train(
    model, X_train, y_train, optimizer=None
)
```

```python
_, _, _, _ = FMM.evaluate(model, X_test, y_test, history)
FMM.plot_history(history)
```

## Using the second version

```python
model = FMM.create_model_v2()
history = FMM.compile_and_train(
    model, X_train, y_train, optimizer=None
)
```
